# 背景知识

面向RAG的大模型应用的核心是搜索索引，它存储并检索向量化内容。查询首先被向量化，与所有块向量进行距离计算。为有效检索上万元素，需采用如faiss、nmslib或annoy等近似最近邻算法的向量索引。同时，ElasticSearch等向量数据库提供数据摄取流水线管理。关于向量数据库部分可以参考：[向量数据库介绍](0.向量数据库介绍.ipynb)

索引就是将文本进行切块处理，放入向量空间中，生成查询关键词，以便后续提问时找到应答参考知识点。具体来讲有以下三种方式

1. **方法一：直接分段**，即将文本按语义细分为多段，每段作为一个向量单位存储，相当于给出的参考不是一本书，而是哪个章节，效率更高。

2. **方法二：生成问答对**，即对方法一的升级，将文本按语义细分为多段，每段作为一个向量单位存储，同时还生成相关的问题和答案以供以后快速响应问题，相对于方法一，不用再次理解段落，直接抄答案就行了。

3. **方法三：增强信息**，即对方法二的升级，不仅备好了问题和答案，而且还会给出更多相关的辅助扩展信息，更为周全和灵活。

索引在RAG技术中，除了增强检索效率和效果外，还有如下几个作用
1. **元数据管理**：索引可以存储与向量相关的元数据，例如文档的标题、作者、创建日期等。这些元数据可以帮助用户更好地理解和筛选检索到的结果。
2. **数据权限控制**：通过索引，可以实现对不同用户的访问控制。根据用户的权限，系统可以过滤掉他们无权访问的数据，从而保护敏感信息。
3. **数据过滤**：索引可以根据特定的条件（如日期范围、数据源等）对数据进行过滤，确保用户获取到最相关和最新的信息。

本部分将围绕一下几个方面来介绍索引相关技术
1. **Metadata和权限体系**：通过管理文档的元数据和用户权限，确保信息的安全性和可控性。具体内容：[Metadata和权限体系](2.Metadata和权限体系.ipynb)
2. **Contextual Chunk Headers** (CCH)：为每个数据块添加上下文头信息，以便于快速识别和处理相关内容。具体内容：[Contextual Chunk Headers](3.CCH_Contextual_Chunk_Headers.ipynb)
3. **Document Augmentation**：通过增强文档内容，提供更多的背景信息和相关数据，以提升检索的准确性和丰富性。具体内容：[Document Augmentation](4.Document_Augmentation.ipynb)
